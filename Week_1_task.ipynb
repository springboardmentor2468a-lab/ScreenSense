```python
print("WEEK - 1")
print("ScreenSense: Kids Screen Time")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("1. Loading Dataset")

df = pd.read_csv("C:\\Users\\meena\\Downloads\\Indian_Kids_Screen_Time.csv") 

print("First 5 rows:")
print(df.head())

print("2. Dataset Overview")

print("\nDataset Shape:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("3. Check Missing Values")
print("\nMissing Values:")
print(df.isnull().sum())

print("4. Age Distribution")

plt.figure()
plt.hist(df["Age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

print("5. Daily Screen Time Distribution")

plt.figure()
plt.hist(df["Avg_Daily_Screen_Time_hr"], bins=10)
plt.title("Daily Screen Time Distribution")
plt.xlabel("Screen Time (Hours)")
plt.ylabel("Frequency")
plt.show()

print("6. Gender Count")
gender_counts = df["Gender"].value_counts()

plt.figure()
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

print("7. Urban vs Rural Distribution")
location_counts = df["Urban_or_Rural"].value_counts()

plt.figure()
plt.bar(location_counts.index, location_counts.values)
plt.title("Urban vs Rural Distribution")
plt.xlabel("Location Type")
plt.ylabel("Count")
plt.show()

print("8. Screen Time by Device Type")
device_avg = df.groupby("Primary_Device")["Avg_Daily_Screen_Time_hr"].mean()

plt.figure()
plt.bar(device_avg.index, device_avg.values)
plt.title("Average Screen Time by Device")
plt.xlabel("Device Type")
plt.ylabel("Average Screen Time (Hours)")
plt.xticks(rotation=45)
plt.show()

print("9. Recommended Limit Compliance")
counts = df["Exceeded_Recommended_Limit"].value_counts()

plt.figure()
plt.pie(counts.values, labels=["Exceeded", "Within Limit"], autopct='%1.1f%%')
plt.title("Screen Time Limit Compliance")
plt.show()

print("11. Correlation Matrix")
correlation = df.corr(numeric_only=True)

print("\nCorrelation Matrix:")
print(correlation)

print("\n\nEDA SUMMARY")
print("------------")

print("1. The dataset contains 9712 records and 8 columns.")

print("2. The age of students ranges from 8 to 18 years, with an average age of around 13 years.")

print("3. The gender distribution is nearly balanced between male and female students.")

print("4. The average daily screen time is around 4 to 5 hours, with some students reaching up to 13+ hours.")

print("5. A significant number of students have exceeded the recommended screen time limit.")

print("6. The Educational to Recreational screen time ratio suggests slightly more time is spent on recreational activities.")

print("7. The dataset contains missing values in the 'Health_Impacts' column, while other columns have complete data.")

print("8. Overall, the dataset shows that high screen time is common among students and may be associated with health issues such as poor sleep and eye strain.")
```

```python
# Check null values
df.info()

# Check null count properly
df.isna().sum()

# Handle categorical null values
df["Gender"] = df["Gender"].fillna("None")
df["Primary_Device"] = df["Primary_Device"].fillna("None")
df["Health_Impacts"] = df["Health_Impacts"].fillna("None")
df["Urban_or_Rural"] = df["Urban_or_Rural"].fillna("None")

# Handle numerical null values (if any)
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Avg_Daily_Screen_Time_hr"] = df["Avg_Daily_Screen_Time_hr"].fillna(df["Avg_Daily_Screen_Time_hr"].mean())
df["Educational_to_Recreational_Ratio"] = df["Educational_to_Recreational_Ratio"].fillna(
    df["Educational_to_Recreational_Ratio"].mean()
)

# Recheck null values
df.isna().sum()
```
