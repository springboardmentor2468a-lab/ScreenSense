## Handle Missing Values
### Finding:

Health_Impacts → ~33% missing

All other columns → 0% missing

```python
df['Health_Impacts'] = df['Health_Impacts'].fillna("Not Reported")
```

```python
df['Gender'] = df['Gender'].str.strip().str.title()
df['Primary_Device'] = df['Primary_Device'].str.strip().str.title()
df['Urban_or_Rural'] = df['Urban_or_Rural'].str.strip().str.title()
```

## Feature Engineering (Derived Fields)

## Age Bands

```python
bins = [5, 9, 12, 15, 18]
labels = ['Child', 'Pre-Teen', 'Early Teen', 'Late Teen']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

## Screen Time Category

```python
def screen_category(x):
    if x < 2:
        return "Low"
    elif x < 4:
        return "Moderate"
    elif x < 6:
        return "High"
    else:
        return "Very High"

df['Screen_Time_Level'] = df['Avg_Daily_Screen_Time_hr'].apply(screen_category)
```

## Device Share Feature

```python
df['Is_Mobile_Device'] = df['Primary_Device'].apply(
    lambda x: 1 if x in ['Smartphone', 'Tablet'] else 0
)
```

```python
def device_group(x):
    if x in ['Smartphone', 'Tablet']:
        return 'Mobile'
    elif x in ['Laptop', 'Desktop']:
        return 'Computer'
    else:
        return 'Other'

df['Device_Category'] = df['Primary_Device'].apply(device_group)
```

## Educational vs Recreational Flag

```python

df['Learning_Heavy_User'] = df['Educational_to_Recreational_Ratio'].apply(
    lambda x: 1 if x >= 0.5 else 0
)
```

## Health Risk Indicator

```python
df['Health_Risk_Flag'] = df['Avg_Daily_Screen_Time_hr'].apply(
    lambda x: 1 if x > 4 else 0
)
```

```python
df['User_Type'] = df['Educational_to_Recreational_Ratio'].apply(
    lambda x: 'Learning-Dominant' if x >= 0.5 else 'Recreation-Dominant'
)
```

## summary

Age segmentation was implemented using derived age bands to enable demographic-level behavioral analysis.
Device usage patterns were captured through device share features, including mobile dependency and device category grouping.
Activity behavior was modeled using the Educational-to-Recreational usage ratio as a proxy indicator of learning-dominant versus recreation-dominant screen usage.
Weekday/weekend behavioral flags could not be derived due to the absence of temporal attributes in the dataset and are identified as a future enhancement if time-based data becomes available.
